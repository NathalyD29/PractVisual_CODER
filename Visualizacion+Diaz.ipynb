```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#1. Carga de archivo y realizar estadisticas descriptivas
df = pd.read_csv("HRDataset_v14.csv") #Importamos el archivo como dataframe con Pandas
df.head() #Imprimimos el comienzo de la tabla para conocer con que información estamos trabajando
```

```python
df.shape #Verificamos el tamaño del dataframe
```

```python
df.info() #Verificamos que información y que tipo de datos estamos trabajando
```

```python
df1 = df.select_dtypes("number") #Con el método select_types genero un df solo con las variables numéricas 
df1.describe() #Calculamos estadisticas descriptivas del conjunto de columnas
```

```python
df2 = df.select_dtypes("object") #Generamos un df con todos los valores categóricos
df2.describe() #Calculamos estadisticas descriptivas del conjunto de columnas
```

```python
#2. Realizar un histograma con los salarios. ¿Qué ragos de salarios son los más populares?
dfSalario = df["Salary"]
dfSalario
```

```python
fig, ax = plt.subplots(figsize=(8, 4)) #Generamos el histograma con la columna de salarios
ax.hist(dfSalario.values.flatten(), bins=10) 
ax.set_title('Histograma de Salarios')
ax.set_xlabel('Intervalos de salarios')
ax.set_ylabel('Frecuencia absoluta')
```

```python
#Revisando la distribución en el histograma podemos concluir que los salarios
#más populares con los que pertenecen al rango entre 50.000 y 100.000
```

```python
#3.Realizar gráfico de violín con los salarios discriminados por géneros
ax = sns.catplot(data=df, kind='violin', x='Sex', y='Salary', split=True)
ax.set(xlabel='Sex',ylabel='Salary',title='Salarios por Género')
```

```python
#4. Graficar la serie de tiempo correspondiente a la fecha de contratación.

plt.figure()
sns.set(rc={'figure.figsize':(30,5)})
ax = sns.lineplot(data=df, x='DateofHire', y='EmpID', linewidth=1)
ax.set(xlabel='Día de contratación', ylabel='Densificación', title='Contratos')
ax.tick_params(axis='x', labelrotation=90)
plt.grid(True)
```

```python
#Grafica de serie de tiempo correspondiente a la fecha de contratación y el salario que obtiene por dicho contrato
plt.figure()
sns.set(rc={'figure.figsize':(20,10)})
ax = sns.lineplot(data=df, x='DateofHire', y='Salary', linewidth=1)
ax.set(xlabel='Día de contratación', ylabel='Salario', title='Contratos vs Salario')
ax.tick_params(axis='x', labelrotation=90)
plt.grid(True)
```
